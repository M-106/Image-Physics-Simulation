This is just an notebook to get some example ray-traces applied on PhysicsGen

```python
import sys
sys.path += [".."] # os.path.abspath("../img-phy-sim")

import img_phy_sim as ips
from img_phy_sim.data import get_image
```

```python
# get image
input_, output_ = get_image(mode='train', variation="sound_reflection", input_type="osm", output_type="complex_only", 
               shuffle=False, return_output=True, as_numpy_array=True)
```

```python
output_.max()
```

```python
ips.img.imshow(input_, size=4)
```

```python
ips.img.imshow(output_, size=4)
```

```python
ips.img.imshow(output_.clip(0, 1), size=4)
```

```python
ips.img.imshow((output_*-1), size=4, cmap="viridis")
```

```python
ips.img.imshow((output_), size=4, cmap="viridis")
```

```python
ray_img = ips.ray_tracing.trace_and_draw_rays(rel_position=(0.5, 0.5), 
                                              img_src=input_,
                                              wall_values=0.0,
                                              directions_in_degree=ips.math.get_linear_degree_range(0, 360, 1))

ips.img.imshow(ray_img, size=4)
```

```python
ray_img = ips.ray_tracing.trace_and_draw_rays(rel_position=(0.5, 0.5), 
                                              img_src=input_,
                                              wall_values=0.0,
                                              directions_in_degree=ips.math.get_linear_degree_range(0, 360, 1),
                                              show_only_reflections=True)

ips.img.imshow(ray_img, size=4)
```

```python
ray_img = ips.ray_tracing.trace_and_draw_rays(rel_position=(0.5, 0.5), 
                                              img_src=input_,
                                              wall_values=0.0,
                                              directions_in_degree=ips.math.get_linear_degree_range(0, 360, 10),
                                              img_background=input_,
                                              ray_value=2)

ips.img.imshow(ray_img, size=4)
```

```python
ray_img = ips.ray_tracing.trace_and_draw_rays(rel_position=(0.5, 0.5), 
                                              img_src=input_,
                                              wall_values=0.0,
                                              directions_in_degree=ips.math.get_linear_degree_range(0, 360, 10),
                                              img_background=input_,
                                              ray_value=2)

ips.img.imshow(ray_img, size=4, cmap="random")
```

```python
ism = ips.ism.compute_reflection_map(
    source_rel=(0.5, 0.5),
    img=input_,
    wall_values=[0],   
    wall_thickness=1,
    max_order=1,
    step_px=1
)

ips.img.imshow(ism, size=4, cmap="viridis")
```
